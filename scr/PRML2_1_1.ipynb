# ベータ分布
二項分布に対する共役事前分布となる分布

\begin{equation*}
Beta(\mu|a,b) =  \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1}\end{equation*}


#### 共役事前分布
尤度関数に事前分布として掛けることで事後分布と同じ形になる。

#### ガンマ関数
\begin{equation*}
\Gamma(x) \equiv  
  \int_0^\infty u^{x-1}e^{-u} dx
\end{equation*}

ガンマ関数の性質

+ $\Gamma(x+1) = x\Gamma(x)$
+ $\Gamma(1) = 1$
+ $x$が正の整数なら$\Gamma(x+1)=x!$


```
# ベータ分布

import math
import numpy as np
import matplotlib.pyplot as plt

params = [(0.5, 0.5), (1., 0.5), (2., 0.5), (3., 0.5), (4., 0.5),
          (0.5, 1.), (1., 1.), (2., 1.), (3., 1.), (4., 1.),
          (0.5, 2.), (1., 2.), (2., 2.), (3., 2.), (4., 2.),
          (0.5, 3.), (1., 3.), (2., 3.), (3., 3.), (4., 3.),
          (0.5, 4.), (1., 4.), (3., 4.), (3., 4.), (4., 4.)
         ]

fig,axes = plt.subplots(nrows=5,ncols=5,figsize=(10,8))
fig.subplots_adjust(wspace=0.4, hspace=0.6)


def func_beta(a, b):
  mus = np.linspace(0, 1, 50)
  C = math.gamma(a + b) / (math.gamma(a) + math.gamma(b))
  beta = [C * mu ** (a - 1) * (1- mu) ** (b - 1) for mu in mus]

  axes[i, j].plot(mu, beta)
  axes[i, j].set_title("a=" + str(a) + ", b=" + str(b))

  
print(type(params[1][0]))
  
for i in range(5):
  for j in range(5):
    func_beta(params[i * 5 + j][0], params[i * 5 + j][1])
```
