# ベータ分布
二項分布に対する共役事前分布となる分布

\begin{equation*}
Beta(\mu|a,b) =  \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1}\end{equation*}


#### 共役事前分布
尤度関数に事前分布として掛けることで事後分布と同じ形になる。

#### ガンマ関数
\begin{equation*}
\Gamma(x) \equiv  
  \int_0^\infty u^{x-1}e^{-u} dx
\end{equation*}

ガンマ関数の性質

+ $\Gamma(x+1) = x\Gamma(x)$
+ $\Gamma(1) = 1$
+ $x$が正の整数なら$\Gamma(x+1)=x!$


```
# ガンマ関数

import math
import numpy as np
import matplotlib.pyplot as plt

num_plots = range(1,8)
f, axarr = plt.subplots(max(num_plots), sharex=True, figsize=(10, 12))
f.suptitle('Sharing X axis')
[axarr[i-1].scatter(np.linspace(0.2, i, 5000), [math.gamma(a) for a in np.linspace(0.2, i, 5000)]) for i in num_plots]
```
